```python
import requests
import bs4
import openpyxl
import os

wb = openpyxl.Workbook()
ws = wb.active

ws["A1"] = "職缺名稱"
ws["B1"] = "職缺連結"
ws["C1"] = "公司名稱"
ws["D1"] = "工作地區"
ws["E1"] = "薪資待遇"
ws["F1"] = "計薪方式"
ws["G1"] = "薪資上限"
ws["H1"] = "薪資下限"


# ws["F1"] = "學歷"

page = 1

res = requests.get("https://www.1111.com.tw/search/job?page=1&col=ab&sort=desc&ks=%E5%A4%A7%E6%95%B8%E6%93%9A",verify=False)
soup = bs4.BeautifulSoup(res.text, "html.parser")

while soup.find_all("div", class_="flex flex-col lg:flex-row lg:gap-4") != []:
    print("===============")
    print("正在抓第", page, "頁")
    print("===============")

    for job in soup.find_-all("div", class_="flex flex-col lg:flex-row lg:gap-4"):
        a = job.h2.text.strip()
        b = "https://www.1111.com.tw" + job.a["href"]#職缺連結
        c = job.find("h2", class_="inline").text.strip()#職缺公司名稱
        d = job.find("div", class_="job-card-condition__item").text.strip()#工作地點
        e = job.h4.text.strip()#薪資待遇
        f = job.find_all("h4", class_="job-card-condition__text")[1].text.strip()#學歷
        # e ="月薪30000元~32000元"
        # e ="月薪33000元以上"
        # e ="時薪 200元~300元上"
        # e ="面議（經常性薪資達4萬元或以上"
        # e = "年薪1,000,000元~1,300,000元"
        # e = "論件計酬 30000元以上"
        # 提取計薪方式
        f=e[:2]
    
        # 篩選字元
        salary=""
        for char in e:
            if char.isdigit() or char=="~":
                salary+=char
        if "~" in salary:
            low = salary[:salary.find("~")]
            high = salary[salary.find("~") + 1:]
        else:
            low = salary
            high = salary
        if "萬" in e:
            low=int(low)*10000
            high=int(high)*10000
        else:
            low=int(low)
            high=int(high)
                
    # print(f,type(f))
    # print(low, type(low))
    # print(high, type(high))

        ws.append([a,b,c,d,e,f,low,high])

    page += 1
    
    res = requests.get("https://www.1111.com.tw/search/job?page=" + str(page) + "&col=ab&sort=desc&ks=%E5%A4%A7%E6%95%B8%E6%93%9A",verify=False)
    soup = bs4.BeautifulSoup(res.text)

wb.save("1111大數據職缺.xlsx")
os.startfile("1111大數據職缺.xlsx")
```
